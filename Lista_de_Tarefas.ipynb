**Listar tarefas pendentes**

**Listar tarefas por categoria**

```python
import json

tarefas = []

def salvar():
  with open("tarefas.json", "w") as f:
    json.dump(tarefas, f, indent=4)

def carregar():
  try:
    with open("tarefas.json", "r") as f:
      data = json.load(f)
      print("Tarefas carregadas com sucesso") # Moved this line to be reachable
      return data
  except FileNotFoundError:
    print("Nenhuma tarefa encontrada")
    return []
  except json.JSONDecodeError: # Handle cases where file exists but is empty or malformed JSON
    print("Arquivo de tarefas encontrado, mas vazio ou corrompido. Iniciando com lista vazia.")
    return []

tarefas = carregar() # Assign the loaded tasks to the global tarefas list

def adicionar():
  nome = input("Digite a descrição da tarefa: ")
  categoria = input("Digite a categoria da tarefa: ")

  tarefa = {
    "nome": nome,
    "categoria": categoria,
    "concluida": False
  }

  tarefas.append(tarefa)
  salvar()
  print("Tarefa adicionada com sucesso")


def listar():
  if not tarefas:
    print("Nenhuma tarefa encontrada")
    return

  print("\n--- Todas as Tarefas ---")
  for i, t in enumerate(tarefas):
    status = "Concluída" if t["concluida"] else "Pendente"
    print(f"{i+1}. Nome: {t['nome']}, Categoria: {t['categoria']}, Status: {status}") # Fixed t[status] to status


def concluir(): # Moved function to global scope
  if not tarefas:
    print("Nenhuma tarefa para concluir.")
    return

  print("\n--- Tarefas disponíveis para Concluir ---")
  for i, t in enumerate(tarefas):
    if not t["concluida"]:
      print(f"{i+1}. Nome: {t['nome']}, Categoria: {t['categoria']}")

  try:
    i = int(input("Digite o número da tarefa a ser marcada como concluída: ")) - 1

    if 0 <= i < len(tarefas):
      tarefas[i]["concluida"] = True
      salvar()
      print("Tarefa concluída com sucesso")
    else:
      print("Número de tarefa inválido")
  except ValueError:
    print("Entrada inválida. Por favor, digite um número.")


def listar_pendentes(): # Moved function to global scope
  pendentes = [t for t in tarefas if not t["concluida"]]
  if not pendentes:
    print("Nenhuma tarefa pendente encontrada")
  else:
    print("\n--- Tarefas Pendentes ---")
    for i, t in enumerate(pendentes):
      print(f"{i+1}. Nome: {t['nome']}, Categoria: {t['categoria']}") # Fixed loop and variable usage


def listar_por_categoria(): # Moved function to global scope
  categoria_input = input("Digite a categoria: ")
  tarefas_categoria = [t for t in tarefas if t["categoria"].lower() == categoria_input.lower()] # Added .lower() for case-insensitive comparison
  if not tarefas_categoria:
    print(f"Nenhuma tarefa encontrada para a categoria '{categoria_input}'")
  else:
    print(f"\n--- Tarefas na categoria '{categoria_input}' ---")
    for i, t in enumerate(tarefas_categoria):
      status = "Concluída" if t["concluida"] else "Pendente"
      print(f"{i+1}. Nome: {t['nome']}, Categoria: {t['categoria']}, Status: {status}") # Fixed t[status] to status and loop


def menu():
  while True:
    print("\n--- MENU ---")
    print("\nEscolha uma opção:")
    print("1. Adicionar tarefa")
    print("2. Listar todas as tarefas")
    print("3. Marcar tarefa como concluída")
    print("4. Listar tarefas pendentes")
    print("5. Listar tarefas por categoria")
    print("0. Sair")

    opc = input('Escolha uma opção: ')
    if opc == '1':
      adicionar()
    elif opc == '2':
      listar()
    elif opc == '3':
      concluir()
    elif opc == '4':
      listar_pendentes()
    elif opc == '5':
      listar_por_categoria()
    elif opc == '0':
      print("Saindo...")
      break
    else:
      print("Opção inválida")

menu()
```
