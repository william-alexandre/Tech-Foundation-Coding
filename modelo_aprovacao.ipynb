***resumo do código em etapas claras:***

**Criação do conjunto de dados**

  É criado um DataFrame com notas (nota 1, nota 2), frequencia e o rótulo aprovado (1 = aprovado, 0 = reprovado).

  O DataFrame é exibido com print(dados).

**Preparação dos dados para o modelo**

  As variáveis independentes (x) são: nota 1, nota 2 e frequencia.

  A variável dependente (y) é aprovado.

**Divisão em treino e teste**

  Os dados são divididos em conjuntos de treino (70%) e teste (30%) usando train_test_split.

**Criação e treinamento do modelo**

É criado um modelo de Regressão Logística.

O modelo é treinado com os dados de treino (x_treino, y_treino).

**Previsão para um novo aluno**

  Um novo caso com valores (nota 1 = 6.5, nota 2 = 7, frequencia = 90) é testado.

  O método predict_proba calcula a probabilidade de aprovação.

**Resultado final**

  O código imprime a probabilidade de aprovação em porcentagem para o novo aluno.

👉 Em resumo, o código cria um modelo de machine learning que aprende com dados de alunos e estima a chance de aprovação de um novo aluno com base em notas e frequência.

```python
import pandas as pd
dados = pd.DataFrame({
    'nota 1' :[5, 6, 7, 8, 3, 4, 9, 10, 6],
    'nota 2' :[6, 5, 8, 9, 2, 3, 10, 9, 6],
    'frequencia': [80, 90, 98, 50, 60, 100, 40, 99, 85],
    'aprovado': [1, 1, 1, 0, 0, 1, 0, 1, 1]
})

print(dados)

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x= dados[['nota 1', 'nota 2', 'frequencia']]
y= dados['aprovado']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)

modelo = LogisticRegression()
modelo.fit(x_treino, y_treino)

teste_novo = [(6.5, 7, 90)]
prob = modelo.predict_proba(teste_novo)
print(f"Probabilidade de aprovação: {prob[0][1]*100: .1f}%")
```
